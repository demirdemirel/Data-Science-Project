```python
from sklearn.tree import DecisionTreeClassifier
naive_model = DecisionTreeClassifier(random_state=42)
param_model = DecisionTreeClassifier(max_depth=10, min_samples_split=15, min_samples_leaf=15, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from math import*
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
def train_and_predict(model, features, target):
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.33, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("accuracy score: %.2f" % accuracy_score(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    (tn, fp, fn, tp,a) = confusion_matrix(y_test, y_pred).ravel()
    print("confusion matrix")
    print("tn, fp, fn, tp")
    print(tn, fp, fn, tp,a)
```

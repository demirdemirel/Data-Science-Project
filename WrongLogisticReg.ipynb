```python
age_dummies = pd.get_dummies(result1['Age'])
occ_dummies = pd.get_dummies(result1['Occupation'])

# Let's take a quick look at the results
age_dummies.head()
```

```python
occ_dummies.head()
```

```python
age_dummies.columns = ['age1','age2','age3','age4','age5','age6','age7']
occ_dummies.columns = ['other', 'academic','artist',' admin', 'college','customer_service','doctor','manager','farmer','homemaker','K_12Student','lawyer','programmer','retired','marketing','Scientist','self_employed','engineer','tradesman','unemployed','writer']
```

```python
X = result1.drop(['Age','Occupation','Genres;','Title'],axis=1)
```

```python
dummies = pd.concat([age_dummies,occ_dummies],axis=1)
```

```python
# Now Concat the X DataFrame with the dummy variables
X = pd.concat([X,dummies],axis=1)

# Preview of Result
X.head()
```

```python
# Set Y as Target class, Had Affair
Y = result1.Gender

# Preview
Y.head()
```

```python
# Flatten array
Y = np.ravel(Y)

# Check result
Y
```

```python
# Math
import math

# Plot imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# Machine Learning Imports
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split

# For evaluating our ML results
from sklearn import metrics

# Dataset Import
import statsmodels.api as sm
```

```python
# Create LogisticRegression model
log_model = LogisticRegression()

# Fit our data
log_model.fit(X,Y)

# Check our accuracy
log_model.score(X,Y)
```
